```python
# Each tweet's retweet count and favorite count should be scraped using the Twitter API (tweepy)
# and stored as JSON data within a file called 'tweet_json.txt'

run = False
# run = True

if run:
    # for this project I created a twitter app called DAND_t2_p3 (https://DAND.com) with the following keys
    consumer_key = 'HIDDEN'
    consumer_secret = 'HIDDEN'
    access_token = 'HIDDEN'
    access_secret = 'HIDDEN'

    # Instantiate API class object
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_secret)

    api = tweepy.API(auth,
                     wait_on_rate_limit=True,
                     wait_on_rate_limit_notify = True)

    # Prepare list of tweet ids to query, start time, failure list
    tweet_ids = tweet_archive.tweet_id.values
    # print(len(tweet_ids))
    start = time.time()
    count = 0
    fails_dict = {}
    # Save each tweet's returns JSON as a new line in a .txt file
    with open(home+'tweet_json.txt', 'w') as outfile:
        # This loop will take 20-30 minutes to run because of Twitter's rate limit
        for tweet_id in tweet_ids:
            count += 1
            print(str(count) + ": " + str(tweet_id))
            try:
                tweet = api.get_status(tweet_id, tweet_mode = 'extended')
                print("Success")
                json.dump(tweet._json, outfile)
                outfile.write('\n')
            except tweepy.TweepError as e:
                print("Fail")
                fails_dict[tweet_id] = e
                pass
    end = time.time()
    print("Complete. Time Elapsed:")
    print(end-start)
else:
    print('not run')
```

Each needs to be assessed for quality (at least 8) and tidyiness (at least 2) issues.

**Poor Quality, or dirty data, is defined by the following criteria:**

1) Completeness (ex: NaN)

2) Validity (ex: negative height, duplicates)

3) Accuracy (ex: valid but wrong)

4) Consistency (ex: NY vs New York)

**Untidy, or messy data, is defined by the following criteria:**

1) Each variable forms a column

2) Each observation forms a row

3) Each type of observable unit forms a table

```python
# tweet_archive (6)
# Define
# name should not include 'a', 'an', 'the'. Recalculate also allowing for "meet x" format
# Search for 'this is ____' or 'meet ____'

# Code
tweet_archive_clean.name = '0'

def fill_blank_extract(x):
    try:
        s = re.search('[Tt]his is (?:a )?(?:an )?(?:the )?([A-Z][a-z]+)', x['text']).group(1)
    except AttributeError:
        try:
            s = re.search('[Mm]eet (?:a )?(?:an )?(?:the )?([A-Z][a-z]+)', x['text']).group(1)
        except AttributeError:
            s = np.nan
    return s

tweet_archive_clean['name'] = tweet_archive_clean.apply(lambda x: fill_blank_extract(x), axis = 1)
```
